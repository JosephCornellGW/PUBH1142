```python
# A list of daily step counts
steps = [4200, 7800, 6500, 9000, 11000, 3000, 8500]

total_steps = 0

for step_count in steps:
    total_steps += step_count

average_steps = total_steps / len(steps)

print("Total weekly steps:", total_steps)
print("Average daily steps:", round(average_steps))
```

```python
print(f"Steps per day: {steps}")
print(f"Max day: {max(steps)}")
print(f"Min day: {min(steps)}")
```

```python
for i, step_count in enumerate(steps, start=1):
    print(f"Day {i}: {step_count} steps")
```

```python
for i, step_count in enumerate(steps, start=1):
    status = "Good day" if step_count >= 7000 else "Low activity"
    print(f"Day {i}: {step_count} steps - {status}")
```
