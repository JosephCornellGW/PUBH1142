> Answer: Generate a loop that prints n^2 for each of the numbers n=1,2,3,4,5

```python
for n in range(1, 6):
    print(n**2)
```

```python
[print(n**2) for n in range(1,6)]; # Alternative using list comprehension
```

*Question 4: Create the variable i by assigning the value 1 to it. Create a loop that prints the value of i^2 to the screen.*

> Answer:


```python
i = 1
while i**2 < 1000:
    i += 1
print(i)
```

At `i=31` the value of `i**2` is `961` and at `i=32` the value of `i**2` is `1024`. The value of `i` is `32` after the loop has completed.
