## Else

```python
variable = 10
```

```python
if variable > 5:
    print("Variable is greater than 5")
else:
    print("Variable is not greater than 5")
```

```python
if variable > 5:
    print("Variable is greater than 5")
elif variable == 5:
    print("Variable is equal to 5")
else:
    print("Variable is less than 5")
```

```python
new_variable = 5
```

```python
if new_variable < 0:
    print("Negative")
elif new_variable < 5:
    print("Positive but less than 5")
elif new_variable == 5:
    print("Equal to 5")
else:
    print("Greater than 5")
```
