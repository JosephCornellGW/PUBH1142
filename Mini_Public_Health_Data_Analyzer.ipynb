# Lists and Basic Stats

## List of my daily step counts for 7 days

```python
steps = [10407, 10869, 11247, 18535, 21974, 8904, 10527, 10123]
```

## Calculate total steps

```python
total_steps = sum(steps)
```

## Calculate Average

```python
average_steps = total_steps / len(steps)
```

```python
print("Total steps:", total_steps)
print("Average steps:", average_steps)
```

## if/elif/else

```python
if average_steps >= 15000:
    print("Excellent activity level")
elif average_steps >= 10000:
    print("Moderate activity level")
else: 
    print("Low activity level")
```
