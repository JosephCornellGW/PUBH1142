## **Quiz Questions**

*Question 1: What is the difference between a spreadsheet and a database?*

> Answer: 
> A spreadsheet (like Excel) is built for organizing and analyzing data in rows and columns, usually on a smaller scale and often by one person at a time. It’s flexible, visual, and calculation-focused. A database is built to store, manage, and retrieve large amounts of structured data efficiently. It’s designed for multiple users, relationships between tables, data integrity, and long-term scalability. Spreadsheets analyze. Databases manage.

*Question 2: What are the advantages and disadvantages of using a spreadsheet for health data management?*

>Answer:
> Advantages
>- Easy to use and widely accessible  
>- Strong for quick analysis and visualization  
>- Flexible for exploratory data work  
>- Suitable for small datasets
> Disadvantages
>- Prone to human error  
>- Limited scalability  
>- Weak data validation controls  
>- Not ideal for multi-user environments  
>- Limited security for sensitive health data  

*Question 3: What are the advantages and disadvantages of using a database for health data management?*

> Answer:
> Advantages
>- Efficiently handles large datasets  
>- Enforces data integrity and validation  
>- Supports multiple concurrent users  
>- Stronger security controls  
>- Suitable for long-term structured storage  
>
> Disadvantages
>- Requires technical knowledge like SQL or schema design  
>- More complex to set up and maintain  
>- Less flexible for quick visual exploration
>
> Databases require planning and structure, but they provide reliability at scale.

*Question 4: What are some examples of tasks that can be performed using a spreadsheet?*

> Answer: 
>- Calculating summary statistics  
>- Creating charts and dashboards  
>- Tracking small clinic patient visits  
>- Budget tracking  
>- Performing basic regression analysis  
>- Conducting quick "what-if" scenario modeling  


*Question 5: What are some examples of tasks that can be performed using a database?*

> Answer:
>- Managing electronic health records  
>- Storing longitudinal patient data  
>- Linking lab results to patient identifiers  
>- Running structured queries across large datasets  
>- Supporting hospital-wide reporting systems

*Question 6: What are some examples of tasks that can be performed using both a spreadsheet and a database?*

> Answer:
>- Storing patient demographic data  
>- Filtering and sorting records  
>- Performing basic statistical calculations  
>- Generating summary reports  
>- Exporting data for further analysis  
>- Cleaning structured datasets
>
> The difference lies in scale, security, and reliability.

*Question 7: What are some examples of tasks that can be performed using neither a spreadsheet nor a database?*

> Answer: 
>- Real-time physiological monitoring  
>- Conducting surgical procedures  
>- MRI image reconstruction  
>- Training complex machine learning models without additional tools
>
> Spreadsheets and databases store and oranize data, but they are not full analytical or clinical systems.

*Question 8: What are some examples of tasks that can be performed using a spreadsheet but not a database?*

> Answer:
>- Interactive cell-by-cell modeling  
>- Drag-and-drop visual formatting  
>- Immediate formula experimentation within individual cells  
>- Rapid manual adjustments for exploratory analysis  
>
>Databases are structured systems and do not operate at the individual cell level.

*Question 9: What are some examples of tasks that can be performed using a database but not a spreadsheet?*

> Answer: 
>- Enforcing relational integrity across multiple linked tables  
>- Handling millions of records efficiently  
>- Supporting high-level concurrent multi-user access  
>- Implementing role-based access control  
>- Performing optimized SQL joins across large datasets  
>
> Spreadsheets become inefficient and unstable at this scale.

*Question 10: What are some examples of tasks that can be performed using both a spreadsheet and a database?*

> Answer:
>- Organizing structured datasets into rows and fields  
>- Filtering records based on specific criteria  
>- Calculating totals, averages, and counts  
>- Generating basic reports  
>- Exporting data into other formats 
>- Performing simple data cleaning operations  
>
>Both tools can perform these tasks, but databases are better suited for large, complex, and multi-user environments, while spreadsheets are more practical for smaller, exploratory work.
