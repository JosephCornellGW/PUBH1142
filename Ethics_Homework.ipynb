## **Essay Tasks**

### **The Misuse of Health Data**

*Task: Find another example of the misuse of health data. Give details about the event and discuss the ethical issues involved. What could have been done to prevent this from happening?*

> Answer: An example of the misuse of health data is the case of Henrietta Lacks. In 1951, cancer cells were taken from Lacks during treatment for cervical cancer without her knowledge or consent, and these cells later known as HeLa cells were widely used in medical research and contributed to major scientific advances. Neither Lacks nor her family were informed at the time, and her identity was later publicly disclosed, raising concerns about privacy and exploitation. The ethical issues in this case include the lack of informed consent and respect for patient autonomy, particularly involving a vulnerable patient population. This situation could have been prevented through clear consent processes, transparency, and ethical oversight, which are now central principles in modern health data and research ethics.

### **Breaches in Health Data Security**

*Task: Find another example of a breach in health data. Describe the event, including the number of people affected and the consequences. What could have been done to prevent this from happening?*

> Answer: An example of a breach in health data security that I was affected by is the 2011 Tricare data breach involving the United States military health insurance provider. In this incident, backup tapes containing personal and health information of approximately five million patients were stolen from a vehicle during transport between facilities. Although it was unclear whether the data could be accessed or decrypted, the loss of the tapes was still considered a breach because sensitive health information was taken without authorization. The ethical concerns center on data protection and responsibility in handling patient information. This breach could have been prevented through stronger safeguards, including minimizing physical transport of data, ensuring secure storage during transit, and implementing stricter data handling and oversight procedures.

### **Bias in Health Data**

*Task: Find another example of bias in health data. Describe the event and the consequences. What could have been done to prevent this from happening?*

> Answer: A well known example of bias in health data is the historical underrepresentation of women in clinical trials. For many years, medical research studies primarily enrolled male participants, and the resulting data were used to guide diagnosis, dosing, and treatment for the general population. This imbalance led to gaps in understanding how diseases and medications affect women differently, sometimes resulting in ineffective treatment or increased risk of adverse effects. The consequences of this bias include delayed recognition of gender specific symptoms and inappropriate medication dosing. This issue could have been prevented through inclusive study design, balanced participant recruitment, and earlier requirements for sex based analysis in clinical research.

### **Anonymization of Health Data**

*Task: Find an example of anonymization of health data. Describe the technique used.*

> Answer: An example of anonymization of health data is the use of de identified patient records in public health datasets released under HIPAA guidelines. In these datasets, direct identifiers such as names, addresses, dates of birth, and medical record numbers are removed before the data are shared for research or analysis. This technique, often referred to as data masking and generalization, allows researchers to study trends and outcomes without being able to trace the information back to specific individuals. By removing or generalizing identifying details, patient privacy is protected while still preserving the overall usefulness of the data for population level analysis.

### **Consent in health data collection**

*Task: Find an example of a pro-forma consent document for health data collection.*

>Answer: An example of a pro forma consent document for health data collection is the standard informed consent form used in clinical research studies approved by an institutional review board. These documents typically explain the purpose of the study, what data will be collected, how the data will be used and stored, who will have access to it, and the potential risks and benefits of participation. They also clearly state that participation is voluntary and that individuals may withdraw their consent at any time without penalty. This type of standardized consent document helps ensure transparency, protects participant autonomy, and establishes ethical accountability in health data collection.

## **Quiz Questions**

*Question 1: What is the primary goal of data anonymization?*

> Answer: The primary goal of data anonymization is to protect individual privacy by preventing health data from being linked back to specific people while still allowing the data to be used for research or analysis.

*Question 2: What is the Havasupai Tribe case about?*

> Answer: The Havasupai Tribe case involves researchers using blood samples for purposes beyond what the tribe originally consented to, including genetic studies unrelated to the stated goal, leading to ethical violations involving consent and cultural respect.

*Question 3: What was the largest health data breach in history?*

> Answer: One of the largest health data breaches in history was the 2015 Anthem breach, which exposed personal information of approximately 78 million individuals.

*Question 4: What is the pulse oximeter controversy about?*

> Answer: The pulse oximeter controversy centers on evidence that these devices can produce less accurate oxygen readings in people with darker skin, leading to delayed or missed diagnosis of low oxygen levels.

*Question 5: What is the role of an Institutional Review Board (IRB) in research?*

> Answer: The role of an IRB is to review and oversee research involving human subjects to ensure ethical standards are met and participant rights and safety are protected.

*Question 6: What is the difference between spreadsheet software and database software for health data management?*

> Answer: Spreadsheet software is best for small scale data and simple analysis, while database software is designed to handle large datasets, enforce data integrity, and support secure multi user access.

*Question 7: What is a conflict of interest in public health research?*

> Answer: A conflict of interest occurs when personal, financial, or professional interests have the potential to influence research decisions or outcomes.

*Question 8: What is informed consent in the context of health data collection?*

> Answer: Informed consent means individuals are clearly informed about how their data will be collected, used, stored, and shared, and they voluntarily agree to participate.

*Question 9: What is data masking in the context of de-identification of health data?*

> Answer: Data masking involves removing or replacing identifying information such as names or identification numbers so individuals cannot be easily identified.

*Question 10: What is the principle of k-anonymity in the context of de-identification of health data?*

> Answer: The principle is that individual records are modified so that each person cannot be distinguished from a group of others who share the same characteristics, making it difficult to identify any single individual in the dataset.

*Question 11: What is the main concern about conflicts of interest in public health research?*

> Answer: The main concern is that conflicts of interest can bias research findings and undermine public trust.

*Question 12: What is the first step in managing conflicts of interest in public health research?*

> Answer: The first step is full disclosure of any potential conflicts of interest.

*Question 13: What is the role of independent oversight in managing conflicts of interest in public health research?*

> Answer: Independent oversight helps ensure research decisions are objective and ethical by reviewing conflicts and enforcing safeguards.

*Question 14: What is one strategy for mitigating conflicts of interest in public health research?*

> Answer: One strategy is limiting or separating financial interests from research decision making roles.

*Question 15: What is the potential impact of not properly managing conflicts of interest in public health research?*

> Answer: Failing to manage conflicts of interest can lead to biased results, harmful policy decisions, and loss of public trust.
