# Week 10: Data Visualization with Seaborn

This notebook answers week 10 homework questions

```python
# 1) Import Seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Example DataFrame used for syntax demonstrations
np.random.seed(42)
df = pd.DataFrame({
    "x": np.random.randn(200),
    "y": np.random.randn(200) * 2 + 1,
    "group": np.random.choice(["A", "B", "C"], size=200)
})

df.head()
```

```python
# 2) Scatter plot for x and y in DataFrame df
ax = sns.scatterplot(data=df, x="x", y="y")
plt.show()
```

```python
# 3) Line plot for x and y in DataFrame df
ax = sns.lineplot(data=df, x="x", y="y")
plt.show()
```

```python
# 4) Histogram for x in DataFrame df
ax = sns.histplot(data=df, x="x")
plt.show()
```

```python
# 5) Stacked histogram for x with hue category (using 'group')
ax = sns.histplot(data=df, x="x", hue="group", multiple="stack")
plt.show()
```

```python
# 6) Heatmap for variables x and y (correlation heatmap)
ax = sns.heatmap(df[["x", "y"]].corr(), annot=True, cmap="Blues")
plt.show()
```

```python
# 7) Box plot for group and y in DataFrame df
ax = sns.boxplot(data=df, x="group", y="y")
plt.show()
```

```python
# 8) Violin plot for group and y in DataFrame df
ax = sns.violinplot(data=df, x="group", y="y")
plt.show()
```

```python
# 9) Bar plot for group and y in DataFrame df
ax = sns.barplot(data=df, x="group", y="y")
plt.show()
```

```python
# 10) Add a title to a Seaborn plot assigned to ax
ax = sns.scatterplot(data=df, x="x", y="y")
ax.set_title("My Title")
plt.show()
```

```python
# 11) Save a Seaborn plot to output.png
ax = sns.scatterplot(data=df, x="x", y="y")
ax.figure.savefig("output.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# 12) Set x and y axis labels for ax
ax = sns.scatterplot(data=df, x="x", y="y")
ax.set_xlabel("X Label")
ax.set_ylabel("Y Label")
plt.show()
```

```python
# 13) Change color palette to Blues
sns.set_palette("Blues")
ax = sns.barplot(data=df, x="group", y="y")
plt.show()
```

```python
# 14) Display a Seaborn plot with logarithmic scale on the y-axis
ax = sns.scatterplot(data=df, x="x", y="y")
ax.set_yscale("log")
plt.show()
```

```python
# 15) Add a regression line to a scatter plot for x and y in DataFrame df
ax = sns.regplot(data=df, x="x", y="y")
plt.show()
```
